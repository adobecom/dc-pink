```typescript
import { DOMParser } from "https://deno.land/x/deno_dom/deno-dom-wasm.ts";

function parsePlainHtml(text) {
  return new DOMParser().parseFromString(text, 'text/html');
}

async function readPlainHtml(url) {
  const resp = await fetch(url);
  if (!resp.ok) return null;
  return parsePlainHtml(await resp.text());
}

async function readPlainHtml(url) {
  const resp = await fetch(url);
  if (!resp.ok) return null;
  return parsePlainHtml(await resp.text());
}
```

```typescript
const urlTemplate = 'https://www.adobe.com/acrobat/online/${verb}.html';
const unityVerbs = [
  'sign-pdf',
  'compress-pdf',
  // 1a
  'pdf-editor',
  'add-pdf-page-numbers',
  'crop-pdf',
  'split-pdf',
  // 1b
  'delete-pdf-pages',
  'add-pages-to-pdf',
  'rearrange-pdf',
  'extract-pdf-pages',
  'request-signature'
];
for (const verb of unityVerbs) {
  const url = urlTemplate.replace('${verb}', verb);
  const urlPlainHtml = url.replace('.html', '.plain.html');
  const document = await readPlainHtml(urlPlainHtml);
  if (!document) {
    console.log(`${verb} not found`);
    continue;
  }
  const icons = document.querySelectorAll('span.icon');
  if (icons.length > 0) {
    console.log(`${verb} has span.icon.`);
  }
}
```

```typescript
const urlTemplate = 'https://www.adobe.com/${locale}/acrobat/online/${verb}.html';
const locales = [
  'uk',
  'fr',
  'de',
  'jp',
  'ca',
  'es',
  'it',
  'in',
  'br',
];
const unityVerbs = [
  'sign-pdf',
  'compress-pdf',
  // 1a
  'pdf-editor',
  'add-pdf-page-numbers',
  'crop-pdf',
  'split-pdf',
];
for (const locale of locales) {
  for (const verb of unityVerbs) {
    const url = urlTemplate.replace('${locale}', locale).replace('${verb}', verb);
    const urlPlainHtml = url.replace('.html', '.plain.html');
    const document = await readPlainHtml(urlPlainHtml);
    if (!document) {
      console.log(`${verb} not found`);
      continue;
    }
    const icons = document.querySelectorAll('span.icon');
    if (icons.length > 0) {
      console.log(`${locale} ${verb} has span.icon.`);
    }
  }
}
```
